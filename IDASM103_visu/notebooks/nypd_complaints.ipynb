# Exploration du fichier NYPD_Complaint_Data_Historic

Ce notebook permet d'explorer les donn√©es des fichiers Complaints_2024.

```python
import pandas as pd

df = pd.read_csv('../data/raw/Complaints_2024.csv', low_memory=False)
df.info()
```

```python
df.head()
```

```python
# Removing data without lat/long
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

```python
df = df[df['BORO_NM'] != '(null)']
```

```python
# adding Borough code for cat
df['BORO_CODE'] = df['BORO_NM'].map({
    'MANHATTAN': 1,
    'BRONX': 2,
    'BROOKLYN': 3,
    'QUEENS': 4,
    'STATEN ISLAND': 5
})

df.to_csv('../data/processed/Complaints_2024.csv')
```
